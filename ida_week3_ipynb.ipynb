## 1. numpy

numpy란? 주로 데이터 처리와 수치 계산에 사용되는 라이브러리

**주요 기능**
- 고성능 수치 계산과 수치 해석
- 통계, 선형대수, 행렬 계산
- 벡터, 다차원 배열 계산
- 난수 생성, 푸리에 변환 기능 지원

숫자계산

```python
import numpy as np
```

**배열 생성**

```python
# nd(n차원)array 배열 생성
np.array([1, 2, 3, 4, 5])
```

```python
# 0으로 채워진 배열 생성: np.zeros()
np.zeros((3, 3))  # 0으로 채워진 3x3 배열 생성
```

```python
# 1로 채워진 배열 생성: np.ones()
np.ones((2, 4))  # 1로 채워진 2x4 배열 생성
```

```python
# 범위 지정하여 배열 생성: np.arange()
np.arange(10)   # 0부터 10-1까지
```

```python
# 범위 지정하여 배열 생성: np.arange()
np.arange(0, 10, 2)  # 0부터 10-1까지 2씩 증가하는 배열
```

```python
# 랜덤 배열 생성: np.random.rand()
np.random.rand(3, 3)  # 3x3 랜덤 배열 생성
```

**배열 조작**

```python
# 형상 변경: np.reshape()
arr1 = np.arange(6)

np.reshape(arr1, [2, 3])
```

```python
# 전치(행과 열을 바꿈): np.transpose()
arr1 = np.array([[1, 2, 3], [4, 5, 6]])

np.transpose(arr1)
```

```python
# 배열 병합: np.concatenate()
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[1, 3, 5], [7, 9, 11]])

np.concatenate([arr1, arr2])
```

```python
# np.concatenate() 축 지정
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[1, 3, 5], [7, 9, 11]])

np.concatenate([arr1, arr2], axis=1)
```

**배열 연산**

```python
# numpy 배열 기본연산: 배열 사칙연산
a_arr = np.array([1, 2, 3])
b_arr = np.array([4, 5, 6])
np.add(a_arr, b_arr)
```

```python
a_arr + b_arr
```

```python
# numpy 배열 기본연산: np.sum()함수사용
data1 = np.array([10, 20, 30, 40, 50])
np.sum(data1)
```

```python
# 행렬 곱 (dot product): np.dot()함수 사용
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

np.dot(arr1, arr2)
```

```python
# 브로드캐스팅 (차원 자동 확장)
arr1 = np.array([[1, 2, 3], [4, 5, 6]])  # 2x3
arr2 = np.array([1, 2, 3])  # 1x3

arr1 + arr2  # B가 A의 행 크기에 맞게 자동 확장
```

## 2. pandas

pandas란? 표 형식의 데이터를 다루는 데 특화된 데이터 분석 라이브러리

**주요 기능**

- 누락 데이터(NaN) 처리의 유연한 기능
- 가변적인 사이즈 조절 기능(데이터프레임을 사용한 고차원적인 열을 추가 · 삭제)
- 데이터의 집계 및 변환 기능
- 자동적이며 명확한 데이터 정렬 기능
- 다양한 형식의 데이터 입출력 기능
- 축(axes)의 다중 라벨링(labeling)
- 시계열 데이터 처리 기능
- 레이블 기반의 슬라이싱, 인덱싱, 부분 추출 기능
- 데이터셋에 대한 피보팅과 유연한 형상 변경 기능

**pandas에 적합한 데이터 유형**

- 통계 데이터셋
- 시계열 형식의 데이터
- 데이터베이스 테이블이나 스프레드시트와 같이 여러 유형의 열(column)로 구성된 표 형식의 데이터

**numpy와 pandas 비교**

| 구분 | numpy | pandas |
| :--- | :--- | :--- |
| **핵심 객체** | `ndarray` (N-차원 배열) | `Series`, `DataFrame` (표 형태) |
| **데이터 타입** | 주로 동일한 타입의 수치 데이터 | 서로 다른 타입의 데이터 혼합 가능 (열별) |
| **주요 용도** | 수치 계산, 선형 대수, 행렬 연산 | 데이터 전처리, 통계 분석, 입출력(CSV, Excel) |
| **특징** | 수학적 연산 속도가 매우 빠름 | 색인(Index)와 라벨을 통한 데이터 관리 용이 |

**Series간 산술 연산**

**DataFrame의 산술 연산**

```python
myindex2 = ['강호민', '유재준', '이수진']
mycolumns = ['서울', '부산', '경주']

myframe = DataFrame(np.reshape([10 * i for i in range(1, 10)], (3, 3)),
                     index=myindex2,
                     columns = mycolumns)
myframe
```

```python
# DataFrame과 Series의 산술 연산 가능
# axis = 0: Series의 색인과 DataFrame의 색인 간 연산
myframe.add(myseries, axis=0)
```
