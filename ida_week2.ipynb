````python
mystring = "Introduction to Data Analysis"
mylist = mystring.split()
mylist
````

````python
# ํ์๋ฒ์งธ ๋จ์ด๋ ๋๋ฌธ์๋ก, ์ง์๋ฒ์งธ ๋จ์ด๋ ์๋ฌธ์๋ก ๋ณ๊ฒฝ
for idx in range(len(mylist)):
    if idx % 2 == 0:
        mylist[idx] = mylist[idx].upper()
    else:
        mylist[idx] = mylist[idx].lower()
mylist
````

````python
# join() ํจ์๋ฅผ ์ด์ฉํด ๋ฌธ์์ด ํฉ์น๊ธฐ
result = ",".join(mylist)
result
````
