| 포맷 코드 | 설명|

## 4. 반복문

반복문이란? 프로그램 소스 코드 내에서 특정한 부분의 코드가 반복적으로 수행될 수 있도록 하는 구문

- for문
    ```
    for 변수 in iterable객체:
        반복실행문
    ```

````python
for x in ["a", "b", "c"]:
    print(x)
````

`range()`: 일정한 범위의 정수 시퀀스를 생성하는 함수

| 사용법 | 설명 | 결과 |
|------|------|------|
| `range(a)` | 0부터 `a`개의 정수 | 0, 1, ... `a-1` |
| `range(a, b)` | `a`부터 `b-1`까지 정수 | `a`, `a+1`, ... `b-1` |
| `range(a, b, c)` | `a`부터 `b-1`까지 정수를 `c`간격으로 skip | `a`, `a+c`, ... (`b-1` 이하 정수) |

````python
for i in range(5):
    print(i)
````

````python
for i in range(2, 5):
    print(i)
````

````python
for i in range(3, 11, 2):
    print(i)
````

string도 리스트나 튜플처럼 반복문 적용 가능

````python
for s in "abc":
    print(s)
````

문자열 조작 함수
| 항목 | 설명 |
|------|------|
| `.split(sep)` | 문자열을 분리하여 리스트 생성, `sep`의 기본값은 공백 |
| `.upper()` | 문자열을 모두 대문자로 변환 |
| `.lower()` | 문자열을 모두 소문자로 변환 |
| `.replace(oldstr, newstr)` | 특정 문자열에서 `oldstr`을 `newstr`로 치환 |
| `'joinstr'.join(iterable)` | 반복 가능한 객체(iterable)의 요소를 `joinstr` 문자열로 결합하여 반환 |

````python
mystring = "Introduction to Data Analysis"
mylist = mystring.split()
mylist
````

````python
# 홀수번째 단어는 대문자로, 짝수번째 단어는 소문자로 변경
for idx in range(len(mylist)):
    if idx % 2 == 0:
        mylist[idx] = mylist[idx].upper()
    else:
        mylist[idx] = mylist[idx].lower()
mylist
````

````python
# join() 함수를 이용해 문자열 합치기
result = ",".join(mylist)
result
````

- `lambda()`: 이름 없이 간단한 기능을 한 줄로 정의할 수 있는 익명 함수

일반 함수

````python
def nolambda(x, y):
    return 3 * x + 2 * y
````

````python
x, y = 3, 5
nolambda(x, y)
````

`lambda` 함수

````python
yeslambda = lambda x, y : 3 * x + 2 * y
````

````python
x, y = 3, 5
yeslambda(x, y)
````

````python
price_dict = {
    "apple": 100,
    "banana": 500,
    "watermelon": 300,
    "strawberry": 700,
}
````

````python
sorted(price_dict)
````

````python
sorted(price_dict.items(), key=lambda x: x[0])
````

````python
sorted(price_dict.items(), key=lambda x: x[1])
````
